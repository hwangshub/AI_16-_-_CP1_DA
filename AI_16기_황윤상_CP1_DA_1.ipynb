```python
path = '/content/drive/MyDrive/project/rent/'
forders = os.listdir('/content/drive/MyDrive/project/rent')
df_all = pd.DataFrame()

for file in forders:
  df = pd.read_csv(path + file, encoding='cp949')
  df_all = pd.concat([df_all, df])
  print(df_all.shape)
```

```python
df = df_all.copy()
df.reset_index(inplace=False)
df.info()
```

```python
# 결측치 확인
df.isnull().sum()
```

```python
# 이상값 삭제
del_nodata = df[df['반납대여소번호'].str.contains('N')].index  # 38906개
df.drop(index=del_nodata, inplace = True)
```

```python
# 생년 데이터 정리
df['생년정리'] = df['생년'].str.extract(r'([0-9]+)')
df['생년정리'] = df['생년정리'].fillna(0)
```

```python
# 성별 데이터 정리
# df[df['성별'].str.contains('N').fillna(False)]  # 2853254개 #확인시 데이터 불일치??
# index_sex = df[df['성별'].str.contains('N').fillna(False)].index
# df[df['성별'].str.contains('N').fillna(False)]['성별'] = 99999   # 안됨

df['성별정리'] = (df['성별'] == 'M')*1 + (df['성별'] == 'F')*2
```

```python
# 사용하지 않을 데이터 제거
df.drop(['대여거치대', '반납거치대', '생년', '성별', '대여대여소ID', '반납대여소ID'], axis=1, inplace=True)
```

```python
# # 데이터형식 변환
# df['이용시간(분)'] = df['이용시간(분)'].str.extract(r'([0-9]+)')
# df['이용시간(분)'] = df['이용시간(분)'].apply(pd.to_numeric)
# df['이용시간(분)'].fillna(0)
# df = df.astype({'대여거치대' : int,'반납거치대' : int, '이용시간(분)': int, '이용거리(M)' : np.float})

df['반납대여소번호'] = df['반납대여소번호'].astype(int)
```

```python
# 시간값 수정하기 https://stackoverflow.com/questions/55440145/pd-to-datetime-valueerror-given-date-string-not-likely-a-datetime

# df['대여일시'] = pd.to_datetime(df['대여일시'], format='%Y-%m-%d %H:%M:%S', errors='raise')
# df['반납일시'] = pd.to_datetime(df['반납일시'], format='%Y-%m-%d %H:%M:%S', errors='raise')

df['대여일시'] = pd.to_datetime(df['대여일시'], errors='coerce')
df['반납일시'] = pd.to_datetime(df['반납일시'], errors='coerce')
```

```python
# 시간추출 https://ponyozzang.tistory.com/689   https://sarah0518.tistory.com/61

df['대여일'] = df['대여일시'].dt.date
df['대여일'] = pd.to_datetime(df['대여일'], format='%Y-%m-%d', errors='raise')
# df['대여시간'] = df['대여일시'].dt.time
# df['대여시간'] = pd.to_datetime(df['대여시간'], format='%H:%M:%S', errors='raise')
df['대여시'] = df['대여일시'].dt.hour
# df['대여분'] = df['대여일시'].dt.minute

df['반납일'] = df['반납일시'].dt.date
df['반납일'] = pd.to_datetime(df['반납일'], format='%Y-%m-%d', errors='raise')
# df['반납시간'] = df['반납일시'].dt.time
# df['반납시간'] = pd.to_datetime(df['반납시간'], format='%H:%M:%S', errors='raise')
df['반납시'] = df['반납일시'].dt.hour
# df['반납분'] = df['반납일시'].dt.minute
```

```python
# 요일 추가  # https://beneagain.tistory.com/43
df['week'] = df['대여일'].dt.dayofweek
# 0 월요일, 1 화요일, 2 수요일, 3 목요일, 4 금요일, 5 토요일, 6 일요일
```

```python
# 대여대여소명
df_total = pd.merge(df_inout, df_nameinfo, how='left', left_on = '대여 대여소번호', right_on = '대여소번호')

df_total.columns = ['rent_num', 'return_num', 'use_time', 'use_distance', 'user', 'use_age', 'use_sex', 'rent_day', 'rent_time', 'return_hour', 'return_hour', 'week',
                    'rent_num2', 'rent_city', 'rent_lat', 'rent_long', 'rent_bikenum']
df_total.drop(['rent_num2'], axis=1, inplace=True)
df_total
```

```python
# 반납대여소명
df_total2 = pd.merge(df_total, df_nameinfo, how='left', left_on = 'return_num', right_on = '대여소번호')

df_total2.columns = ['rent_num', 'return_num', 'use_time', 'use_distance', 'user', 'use_age', 'use_sex', 'rent_day', 'rent_time', 'return_hour', 'return_hour', 'week',
                    'rent_city', 'rent_lat', 'rent_long', 'rent_bikenum',
                    'rent_num3' ,'return_city', 'return_lat', 'return_long', 'return_bikenum']
df_total2.drop(['rent_num3'], axis=1, inplace=True)
df_total2
```
